```python
#2)	Write a Python Program and perform the following operations ?
#I.	Create a List {1225, 4986, 6789, 7890, 2345, 6783, 0987, 1234, 8765, 3456}

List1= [1225, 4986, 6789, 7890, 2345, 6783, 987, 1234, 8765, 3456]

print("list=",List1)
```

```python
# Iterating using a simple for loop
print("Iteration using for loop:")
for element in List1:
    print(element)
```

```python
# Iterating using for loop and range
print("\nIteration using for loop and range:")
for i in range(len(List1)):
    print(List1[i])
```

```python
# Using list comprehension to iterate and print elements
print("\nIteration using List Comprehension:")
[print(element) for element in List1]
```

```python
# Using enumerate to get both index and element
print("\nIteration using enumerate (index and value):")
for index, element in enumerate(List1):
    print(f"Index: {index}, Value: {element}")
```

```python
# Using iter and next functions
print("\nIteration using iter and next functions:")
my_iter = iter(List1)  # Create an iterator

while True:
    try:
        element = next(my_iter)
        print(element)
    except StopIteration:
        break  # Exit the loop when all elements are iterated
```

```python
# Using map function to convert elements to string and print them
print("\nIteration using map function:")
list(map(lambda x: print(x), List1))
```

```python
# Using zip to combine elements with their index or another list
print("\nIteration using zip (list with range of indices):")
for index, element in zip(range(len(List1)), List1):
    print(f"Index: {index}, Value: {element}")
```

```python
import numpy as np

# Converting the list to a NumPy array
np_array = np.array(List1)

print("\nIteration using NumPy array:")
for element in np_array:
    print(element)
```

```python
#3)	For a List A, B, C, D, E write a python program to compute all the combinations and permutations 
# Function to compute permutations
def permutations(elements, r):
    def permute(current, remaining):
        if len(current) == r:
            result.append(current)
            return
        for i in range(len(remaining)):
            permute(current + [remaining[i]], remaining[:i] + remaining[i+1:])
    
    result = []
    permute([], elements)
    return result

# Function to compute combinations
def combinations(elements, r):
    def combine(start, current):
        if len(current) == r:
            result.append(current)
            return
        for i in range(start, len(elements)):
            combine(i + 1, current + [elements[i]])
    
    result = []
    combine(0, [])
    return result

# Define lists A, B, C, D, E
A = [1, 2]
B = [3, 4]
C = [5, 6]
D = [7, 8]
E = [9, 10]

# Combine lists A, B, C, D, E into one single list
single_list = A + B + C + D + E

# Compute permutations (choosing 2 items at a time)
print("Permutations of 2 elements:")
perms = permutations(single_list, 3)
for perm in perms:
    print(perm)

# Compute combinations (choosing 3 items at a time)
print("\nCombinations of 2 elements:")
combs = combinations(single_list, 2)
for comb in combs:
    print(comb)
```

```python
#4)	Using the same list use itertools compute permutations and combinations 
import itertools

# Define lists a, b, c, d, e
a = [1, 2]
b = [3, 4]
c = [5, 6]
d = [7, 8]
e = [9, 10]

# Combine lists a, b, c, d, e into one list of lists
list1 = [a, b, c, d, e]

# Flatten list1 into a single list
single_list = [item for sublist in list1 for item in sublist]

# Compute permutations (choosing 2 items at a time)
print("Permutations of 2 elements:")
permutations = itertools.permutations(single_list, 2)
for perm in permutations:
    print(perm)

# Compute combinations (choosing 2 items at a time)
print("\nCombinations of 2 elements:")
combinations = itertools.combinations(single_list, 2)
for comb in combinations:
    print(comb)
```
